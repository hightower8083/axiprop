Field equations:

$$ \nabla \times \mathbf{E} = - \partial_t \mathbf{B} $$
$$ \nabla \times \mathbf{B} = \frac{1}{c^2} \; \partial_t \mathbf{E} + \mu_0 \mathbf{J}$$

Combine for electric field:

$$ \nabla \times \nabla \times \mathbf{E} = \nabla (\nabla \mathbf{E}) - \nabla^2 \mathbf{E} = - \partial_t \left( \frac{1}{c^2} \; \partial_t \mathbf{E} + \mu_0 \mathbf{J}\right) = - \frac{1}{c^2} \; \partial_t^2 \mathbf{E} - \mu_0  \partial_t\mathbf{J}$$

Negelect the charge density perturbations (plasma waves): $\nabla (\nabla \mathbf{E}) = 0$:

$$ \nabla^2 \mathbf{E} - \frac{1}{c^2} \; \partial_t^2 \mathbf{E} - \mu_0  \partial_t\mathbf{J} = 0$$

Go to RZ geometry and consider only polarisation component:

___
$$ \partial_z^2 E = \frac{1}{c^2} \; \partial_t^2 E - \nabla_\perp^2 E + \mu_0  \partial_t J$$
___

Cosider current derivative term in homogeneous electron plasma with non-relativistic motion equation discarding the magnetic term ($v_z\to0$):

$$ \mu_0  \partial_t J_h = \mu_0  \partial_t (-e n_e v_e) =  -\frac{e n_e}{m_e}  \mu_0  \partial_t m_m v_e =  \frac{e^2 n_e}{m_e\epsilon_0}  \mu_0 \epsilon_0 E = \frac{\omega_{pe}^2}{c^2} E $$

where $ \omega_{pe} = \sqrt{\frac{e^2 n_e}{m_e\epsilon_0}}$. 

and consider full current as $J = J_h + \delta J$ : 

$$ \partial_z^2 E = \frac{1}{c^2} \; \partial_t^2 E - \nabla_\perp^2 E +  \frac{\omega_{pe}^2}{c^2} E + \Upsilon $$
with $ \Upsilon = \mu_0  \partial_t \delta J$.

Go to Fourier space (field will belong to negative frequency part, e.g. $E\propto \exp(-i\omega t)$):
___
$$ \partial_z^2 \hat{E} = - k_\|^2 \hat{E} + \hat{\Upsilon} $$
___
with $k_\| = \sqrt{ k^2 - k_\perp^2 - k_{pe}^2}$, where $k=\omega/c$ and $ k_{pe} = \omega_{pe} / c$

Homogeneous solution of interest (for positive propagation diraction $ E\propto \exp[i (k_\| z-\omega t)]$) is $\hat{E}(z) = \hat{E}_0 \exp(ik_\| z)$. 

1) For inhomogeneous equation we set $\hat{E}(z) = \hat{E}_1 \exp(ik_\| z) + \hat{E}_2$, with $E_2=$const.

From propagation equaiton we get
$$ \hat{E}_2  = k_\|^{-2} \hat{\Upsilon} $$
and from initial conditions:
$$\hat{E}_1 = \hat{E}_0 - k_\|^{-2} \hat{\Upsilon}  $$

So solution reads:
___
$$\hat{E}(z) = \left(\hat{E}_0  - k_\|^{-2} \hat{\Upsilon} \right) \exp(ik_\| z)  + k_\|^{-2} \hat{\Upsilon} $$
___


2) Since we only know current at moment $t=0$, we may take into account that it co-propagates with the field, e.g. homogeneous equation solution, so we add the propagating phase to the source term:
$$ \partial_z^2 \hat{E} = - k_\|^2 \hat{E} + \hat{\Upsilon}\exp(ik_\| z)  $$

For this inhomogeneous equation we set $\hat{E}(z) = \left(\hat{E}_0 + z \hat{E}_2\right) \exp(ik_\| z) $, with $E_2=$const.
From propagation equaiton we get
$$ 2 ik_\| \hat{E}_2 = \hat{\Upsilon} \qquad \to \qquad \hat{E}_2 = - \frac{i}{2 k_\|} \hat{\Upsilon} $$

So solution reads:
___
$$\hat{E}(z) = \left(\hat{E}_0 - \frac{iz}{2 k_\|} \hat{\Upsilon} \right) \exp(ik_\| z)$$
___


#### Altrenatively
we set $\hat{E}(z) = \hat{E}^*(z) \exp(ik_\| z) $, with $E_2=$const.
From propagation equaiton we get envelope propagation
$$ \partial_z^2 \hat{E}^* + 2 ik_\| \partial_z\hat{E}^*  = \hat{\Upsilon} $$
for which we can look for a solution $\hat{E}^* =  \hat{E}^{**} \exp(i\alpha k_\| z)$ and get:
$$ (- \alpha^2 k_\|^2  - 2 \alpha k_\|^2 ) \hat{E}^{**} = \hat{\Upsilon} \qquad\to\qquad 
\hat{E}^{**} =  - \frac{\hat{\Upsilon}}{k_\|^2 (\alpha^2+2\alpha) }$$

and from initial condition $\hat{E}^{**}(0)=\hat{E}_0$:
$$ \alpha^2 + 2\alpha + \frac{\hat{\Upsilon}}{k_\|^2 \hat{E}_0} = 0 $$
so that a smallest solution is $\alpha =  - 1 + \sqrt{ 1 -  \frac{\hat{\Upsilon}}{k_\|^2 \hat{E}_0} } $


And this gives a capacity-like solution:
$$ \hat{E}(z) =  \hat{E}_0 \exp\left(i  k_\| z \sqrt{ 1 -  \frac{\hat{\Upsilon}}{k_\|^2\hat{E}_0} } \right)  $$

we may consider a first order expansion in phase:
$$ \hat{E}(z) = \hat{E}_0 \exp\left[i z \left( k_\| - \frac{\hat{\Upsilon}}{2 k_\|\hat{E}_0}\right) \right] =  \hat{E}_0 \exp\left(i  k_\| z \right)  \exp\left(-i \frac{z}{2 k_\|\hat{E}_0}\hat{\Upsilon} \right)  $$
and if we expand the exponent we'll come to solution above. 

Let us consider more terms:

$$ \hat{E}(z) =  \exp(i  k_\| z )  \left( \hat{E}_0  - i \frac{z}{2 k_\|}\hat{\Upsilon} - \hat{E}_0  \left(\frac{z}{2 k_\|\hat{E}_0}\hat{\Upsilon}\right)^2 \right) $$
___
